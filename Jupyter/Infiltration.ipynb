```python
"""
Hawks and doves can infiltrate each other's populations.
In each case there is a chance of instance death if the invading species can't find any food, as
the environment is already near carrying capacity.
Doves are slightly less successful as they can also be killed be hawks before gaining a foothold.
"""
starting_population = {
    "Hawk": [1],
}

infiltration_dict = {
    "Dove": [1]
}

colors = ['orange','blue']
simulation_results = run_simulation(num_generations=100,starting_population=starting_population,ageing_rate=0.01,
                                    infiltration_dict=infiltration_dict)        
display_results(simulation_results,colors)
```

```python
starting_population = {
    "Dove": [1],
}
infiltration_dict = {
    "Hawk": [1]
}
colors = ['blue','orange']
simulation_results = run_simulation(num_generations=200,starting_population=starting_population,ageing_rate=0.01,
                                    infiltration_dict=infiltration_dict)        
display_results(simulation_results,colors)
```

```python
"""
When geese infiltrate, they always eliminate the hawks.
"""
starting_population = {
    "Hawk": [1],
}
infiltration_dict = {
    "Goose": [1]
}
colors = ['orange','green']
simulation_results = run_simulation(num_generations=200,starting_population=starting_population,ageing_rate=0.01,
                                    infiltration_dict=infiltration_dict)        
display_results(simulation_results,colors)
```

```python
"""
Even with a mass hawk invasion early on, geese are able to eliminate the invaders after suffering a loss
"""
starting_population = {
    "Goose": [1],
}
infiltration_dict = {
    "Hawk": [500]
}
colors = ['green','orange']
simulation_results = run_simulation(num_generations=200,starting_population=starting_population,ageing_rate=0.01,
                                    infiltration_dict=infiltration_dict,infiltration_gen=10)        
display_results(simulation_results,colors)
```

```python
"""
Crows always eliminate doves when introduced
"""
starting_population = {
    "Dove": [1],
}
infiltration_dict = {
    "Crow": [1]
}
colors = ['blue','purple']
simulation_results = run_simulation(num_generations=200,starting_population=starting_population,ageing_rate=0.01,
                                    infiltration_dict=infiltration_dict,infiltration_gen=100)        
display_results(simulation_results,colors)
```

```python
"""
When doves invade geese they coexist, and vice-versa
"""
starting_population = {
    "Goose": [1],
}
infiltration_dict = {
    "Dove": [100]
}
colors = ['green','blue']
simulation_results = run_simulation(num_generations=200,starting_population=starting_population,ageing_rate=0.01,
                                    infiltration_dict=infiltration_dict)        
display_results(simulation_results,colors)
```

```python
"""
When enough hawks invade a stable population of geese and doves, they have the chance to eliminate the geese
by exploiting interactions with doves
"""
starting_population = {
    "Goose": [1],
    "Dove": [1]
}
infiltration_dict = {
    "Hawk": [100]
}
colors = ['green','blue','orange']
simulation_results = run_simulation(num_generations=200,starting_population=starting_population,ageing_rate=0.01,
                                    infiltration_dict=infiltration_dict,infiltration_gen=50)        
display_results(simulation_results,colors)
```

```python
"""
Geese, however, are unable to infiltrate a stable H/D ecosystem
"""
starting_population = {
    "Dove": [1],
    "Hawk": [1]
}
infiltration_dict = {
    "Goose": [100]
}
colors = ['blue','orange','green']
simulation_results = run_simulation(num_generations=200,starting_population=starting_population,ageing_rate=0.01,
                                    infiltration_dict=infiltration_dict,infiltration_gen=50)        
display_results(simulation_results,colors)
```

```python
"""
Hawks can be brought back from the verge of extinction if doves are introduced for them to exploit
"""
starting_population = {
    "Goose": [1],
    "Hawk": [1]
}
infiltration_dict = {
    "Dove": [50]
}
colors = ['green','orange','blue']
simulation_results = run_simulation(num_generations=200,starting_population=starting_population,ageing_rate=0.01,
                                    infiltration_dict=infiltration_dict,infiltration_gen=7)        
display_results(simulation_results,colors)
```

```python
"""
A single crow can upset a stable H/D ecosystem by eliminating doves and reaching its own H/D eq. with the hawks
"""
starting_population = {
    "Dove": [1],
    "Hawk": [1]
}
infiltration_dict = {
    "Crow": [1]
}
colors = ['blue','orange','purple']
simulation_results = run_simulation(num_generations=200,starting_population=starting_population,ageing_rate=0.01,
                                    infiltration_dict=infiltration_dict,infiltration_gen=50)        
display_results(simulation_results,colors)
```
