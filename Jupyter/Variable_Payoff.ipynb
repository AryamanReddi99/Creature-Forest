```python
import numpy as np
import matplotlib.pyplot as plt
import import_ipynb
from ipywidgets import interact
from Creature_Forest_Base import *
```

```python
def plot_expectance(a=1,b=1.5,c=0.5,d=0):
    f = np.linspace(0,1,2)
    plt.xlabel('$fraction$' + ' ' + '$of$' + ' ' + 'A')
    plt.ylabel('$Expectance$')
    plt.title(r"$a$ = {}, $b$ = {}, $c$ = {}, $d$ = {}".format(a,b,c,d))
    plt.plot(f, c + f*(a-c), label = 'A')
    plt.plot(f, d + f*(b-d), label = 'B')
    plt.legend(loc = 'upper left')
    plt.grid()
    plt.show()
interact(plot_expectance, a = (0,2,0.25), b = (0,2,0.25), c = (0,2,0.25), d = (0,2,0.25));
```

```python
payoff_dict = {
    "a": 1,
    "b": 1.5,
    "c": .5,
    "d": 0.25
}

starting_population = {
    "A": [1],
    "B": [1]
}
colors = ['firebrick','goldenrod']
simulation_results = run_simulation(starting_population,ageing_rate=0.0,num_generations=100,payoff_dict=payoff_dict)
display_results(simulation_results,colors)
```
